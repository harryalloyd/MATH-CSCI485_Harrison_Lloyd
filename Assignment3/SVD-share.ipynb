## 1. Example Singular Value Decomposition (SVD) Process:

### Example Matrix
Let’s decompose the matrix $ A $:
\$
A = 
\begin{bmatrix}
3 & 2 \\
2 & 3
\end{bmatrix}
\$

### Step 1: Compute $ U $, $ \Sigma $, and $ V^T $
Using SVD, $ A $ can be expressed as:
> \$
A = U \Sigma V^T
\$
Where:
- $ U $ is the matrix of **left singular vectors** (orthogonal, size $ 2 \times 2 $).
- $ \Sigma $ is the diagonal matrix of **singular values** (size $ 2 \times 2 $).
- $ V^T $ is the transpose of the matrix of **right singular vectors** (orthogonal, size $ 2 \times 2 $).

### Step 2: Calculate Singular Values ($ \Sigma $)

The singular values are the square roots of the eigenvalues of $ A^T A $.  

> $$
A^T A = 
\begin{bmatrix}
3 & 2 \\
2 & 3
\end{bmatrix}
\begin{bmatrix}
3 & 2 \\
2 & 3
\end{bmatrix}
=
\begin{bmatrix}
13 & 12 \\
12 & 13
\end{bmatrix}
$$

The eigenvalues of $ A^T A $ are found by solving $ \text{det}(A^T A - \lambda I) = 0 $:
> \$
\text{det}
\begin{bmatrix}
13 - \lambda & 12 \\
12 & 13 - \lambda
\end{bmatrix}
= (13 - \lambda)^2 - 144 = 0
\$

Solve for $ \lambda $:
> \$
\lambda^2 - 26\lambda + 25 = 0 \quad \Rightarrow \quad \lambda = 25, 1
\$

The singular values are $ \sigma_1 = \sqrt{25} = 5 $ and $ \sigma_2 = \sqrt{1} = 1 $.  
Thus, $ \Sigma = \text{diag}(5, 1) $.

### Step 3: Calculate Right Singular Vectors ($ V $)
The right singular vectors are the eigenvectors of $ A^T A $:
> \$
\begin{bmatrix}
13 & 12 \\
12 & 13
\end{bmatrix}
v = \lambda v
\$

For $ \lambda = 25 $:
> \$
\begin{bmatrix}
13 - 25 & 12 \\
12 & 13 - 25
\end{bmatrix}
v = 
\begin{bmatrix}
-12 & 12 \\
12 & -12
\end{bmatrix}
v = 0
\$

Solve to find $ v_1 = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix} $.

For $ \lambda = 1 $:
> \$
\begin{bmatrix}
13 - 1 & 12 \\
12 & 13 - 1
\end{bmatrix}
v = 
\begin{bmatrix}
12 & 12 \\
12 & 12
\end{bmatrix}
v = 0
\$

Solve to find $ v_2 = \frac{1}{\sqrt{2}} \begin{bmatrix} -1 \\ 1 \end{bmatrix} $.

Thus, $ V = \begin{bmatrix} \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \end{bmatrix} $.

### Step 4: Calculate Left Singular Vectors ($ U $)
The left singular vectors are computed as:
> \$
u_i = \frac{Av_i}{\sigma_i}
\$

For $ v_1 = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix} $:
> \$
u_1 = \frac{A v_1}{\sigma_1} = \frac{1}{\sqrt{2}} \frac{\begin{bmatrix} 3 & 2 \\ 2 & 3 \end{bmatrix} \begin{bmatrix} 1 \\ 1 \end{bmatrix}}{5} = \frac{1}{\sqrt{2}} \frac{\begin{bmatrix} 5 \\ 5 \end{bmatrix}}{5} = \begin{bmatrix} \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} \end{bmatrix}
\$

For $ v_2 = \frac{1}{\sqrt{2}} \begin{bmatrix} -1 \\ 1 \end{bmatrix} $:
> \$
u_2 = \frac{A v_2}{\sigma_2} = \frac{1}{\sqrt{2}} \frac{\begin{bmatrix} 3 & 2 \\ 2 & 3 \end{bmatrix} \begin{bmatrix} -1 \\ 1 \end{bmatrix}}{1} = \frac{1}{\sqrt{2}} \frac{\begin{bmatrix} -1 \\ 1 \end{bmatrix}}{1} = \begin{bmatrix} -\frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} \end{bmatrix}
\$

Thus, $ U = \begin{bmatrix} \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \end{bmatrix} $.

### Final Decomposition
> \$
A = 
\begin{bmatrix}
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \\
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}}
\end{bmatrix}
\begin{bmatrix}
5 & 0 \\
0 & 1
\end{bmatrix}
\begin{bmatrix}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\
-\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}}
\end{bmatrix}
\$

## 2. Key Concept of Dimensionality Reduction and Feature Extraction with SVD

Let's revisit the SVD of the $ 2 \times 2 $ matrix $ A $:
> \$
A = 
\begin{bmatrix}
3 & 2 \\
2 & 3
\end{bmatrix}
\$
Its SVD is:
> \$
A = U \Sigma V^T
\$
where:
- $ U = \begin{bmatrix} \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \end{bmatrix} $
- $ \Sigma = \begin{bmatrix} 5 & 0 \\ 0 & 1 \end{bmatrix} $
- $ V^T = \begin{bmatrix} \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \end{bmatrix} $

Now, let’s break down how SVD enables dimensionality reduction and reveals key features:

---

### **Singular Values Identify Dominant Features**
The diagonal entries of $ \Sigma $ ($ 5 $ and $ 1 $) represent the "importance" or "energy" of the corresponding singular vectors in $ U $ and $ V $. 

- The larger singular value ($ 5 $) corresponds to the direction (principal component) where the majority of the variance or "information" in the data is concentrated.
- The smaller singular value ($ 1 $) corresponds to a less significant direction.

### **Dimension Reduction by Ignoring Less Significant Directions**
To reduce dimensionality:
- Keep only the dominant singular value and its corresponding singular vectors (first row/column of $ \Sigma, U, V $).
- Approximate $ A $ as:
  > \$
  A' = u_1 \sigma_1 v_1^T = \begin{bmatrix} \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} \end{bmatrix} 5 \begin{bmatrix} \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \end{bmatrix}
  \$
  Simplifying:
  > \$
  A' =
  \begin{bmatrix}
  2.5 & 2.5 \\
  2.5 & 2.5
  \end{bmatrix}
  \$

- This $ A' $ is a rank-1 approximation of $ A $, capturing the primary direction of variation while discarding the less important dimension.
- **What does this $A'$ represent**?

### **Revealing Patterns in Data**
In the original 2D space of $ A $, the data is scattered along two correlated axes. The principal component (associated with $ \sigma_1 $) identifies the dominant linear relationship:
- The first right singular vector $ v_1 = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix} $ shows that the data aligns strongly along the direction where both features increase together.
- The second right singular vector $ v_2 = \frac{1}{\sqrt{2}} \begin{bmatrix} -1 \\ 1 \end{bmatrix} $ corresponds to an orthogonal direction, but it contributes much less to the data's variance.

By focusing only on $ v_1 $, SVD reduces the data to its most essential pattern: a 1D line capturing the dominant trend in the original 2D data.

---

### Insights
- **Dimensionality Reduction**: SVD removes noise or redundant directions (those with small singular values), simplifying the data to its most informative components.
- **Feature Extraction**: The dominant singular vectors reveal the principal relationships and trends in the data (e.g., $ v_1 $ shows that the two features are positively correlated).
- **Efficient Representation**: The rank-1 approximation $ A' $ provides a compact representation of $ A $, preserving its essential structure while discarding minor variations.

## 3. Example Application in Image Processing

Let’s blur an image using **SVD**, by keeping only the top $ k $ singular values.

Given a high-resolution image:

- Convert it to grayscale (or do this per RGB channel)
- Apply SVD: \$ A = U \Sigma V^T \$
- Reconstruct using top \$ k \$ singular values → \$ A_k = U_k \Sigma_k V_k^T \$
- Visualize the result as a "blurred" version of the original

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def svd_blur(image_path, k_values=[5, 20, 50, 100]):
    # Load image and convert to grayscale
    img = Image.open(image_path).convert('L')  # 'L' = grayscale
    A = np.array(img, dtype=np.float64)

    # Apply SVD
    U, S, VT = np.linalg.svd(A, full_matrices=False)

    # Plot original
    plt.figure(figsize=(12, 3))
    plt.subplot(1, len(k_values)+1, 1)
    plt.imshow(A, cmap='gray')
    plt.title('Original')
    plt.axis('off')

    # Reconstruct with different k
    for idx, k in enumerate(k_values, start=2):
        Ak = U[:, :k] @ np.diag(S[:k]) @ VT[:k, :]
        plt.subplot(1, len(k_values)+1, idx)
        plt.imshow(Ak, cmap='gray')
        plt.title(f'k = {k}')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Example usage
svd_blur('building.jpg', k_values=[1, 5, 10, 20, 50])
```
